## Deep Learning avec TensorFlow/Keras

```python
import subprocess
import sys

try:
    import tensorflow as tf
    print(f"TensorFlow version: {tf.__version__}")
except ImportError:
    print("Installation de TensorFlow...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "tensorflow"])
    import tensorflow as tf
    print(f"TensorFlow installÃ©: {tf.__version__}")
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

x_train_dense = x_train.toarray()
x_test_dense = x_test.toarray()

print(f"Shape des donnÃ©es d'entraÃ®nement: {x_train_dense.shape}")
print(f"Shape des donnÃ©es de test: {x_test_dense.shape}")
```

```python
model_keras = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(x_train_dense.shape[1],), name='couche_entree'),
    layers.Dropout(0.3, name='dropout_1'),
    
    layers.Dense(64, activation='relu', name='couche_cachee_1'),
    layers.Dropout(0.2, name='dropout_2'),
    
    layers.Dense(32, activation='relu', name='couche_cachee_2'),
    layers.Dropout(0.2, name='dropout_3'),
    
    layers.Dense(1, activation='sigmoid', name='couche_sortie')
])

model_keras.summary()
```

```python
model_keras.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', 'Precision', 'Recall']
)

print("ModÃ¨le compilÃ© et prÃªt pour l'entraÃ®nement!")
```

```python
history = model_keras.fit(
    x_train_dense, 
    y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

print("\nâœ“ EntraÃ®nement terminÃ©!")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Loss (Train)')
plt.plot(history.history['val_loss'], label='Loss (Validation)')
plt.title('Ã‰volution de la fonction de perte')
plt.xlabel('Epoch')
plt.title('Evolution de la loss')
plt.legend()
plt.grid(True)

# Graphique de l'accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Accuracy (Train)')
plt.plot(history.history['val_accuracy'], label='Accuracy (Validation)')
plt.title('Ã‰volution de la prÃ©cision')
plt.xlabel('Epoch')
plt.title('Evolution de l\'accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
test_loss, test_accuracy, test_precision, test_recall = model_keras.evaluate(x_test_dense, y_test, verbose=0)

print("=== RÃ©sultats du RÃ©seau de Neurones (Keras/TensorFlow) ===")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Test Recall: {test_recall:.4f}")
```

### Comparaison finale

```python
comparaison_finale = pd.DataFrame({
    'ModÃ¨le': ['RÃ©gression Logistique', 'MLP (sklearn)', 'Deep Learning (Keras)'],
    'Accuracy': [accuracy_sc, accuracy_nn, test_accuracy],
    'Recall': [recall_sc, recall_nn, test_recall],
    'Precision': [precision_lr, precision_nn, test_precision]
})

print("\n" + "="*70)
print("COMPARAISON FINALE DES 3 MODÃˆLES")
print("="*70)
print(comparaison_finale.to_string(index=False))
print("="*70)

# Identifier le meilleur modÃ¨le
best_idx = comparaison_finale['Accuracy'].idxmax()
best_model = comparaison_finale.loc[best_idx, 'ModÃ¨le']
best_accuracy = comparaison_finale.loc[best_idx, 'Accuracy']

print(f"\nðŸ† MEILLEUR MODÃˆLE: {best_model}")
print(f"   Accuracy: {best_accuracy:.2%}")
```
